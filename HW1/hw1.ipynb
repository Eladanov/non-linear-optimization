```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
data = np.genfromtxt('/spectra.csv',delimiter=',')
Xt = data - np.mean(data, axis = 0)
A = Xt.T @ Xt
u,V= np.linalg.eigh(A)
n=np.size(u)
Xc = Xt @ V[:,[n-2,n-1]]
plt.plot(Xc[:,0],Xc[:,1],"*")
print("proportion of variance by first two components",(u[n-1]+u[n-2]) / sum(u))
```

```python
ax = plt.axes(projection='3d')
Xd = Xt @ V[:,[n-3,n-2,n-1]]
ax.plot3D(Xd[:,0],Xd[:,1],Xd[:,2],"*")
print("proportion of variance by first three components",(u[n-1]+u[n-2]+u[n-3]) / sum(u))
```
